```python
 #CHAT GPT ISHLAB BERDI 

file_path = r"F:\Arslonbek\MySecondBrain\DataAnalitika\Python_homework\lesson-20\homework\chinook.db"

try:
    with sqlite3.connect(file_path) as conn:
        
        # 1. Barcha xarid detallarini (InvoiceItems) albom ma'lumotlari (Albums) bilan birlashtirish
        # Bu so'rov har bir sotib olingan trekka tegishli Albomni aniqlaydi.
        track_data_query = """
        SELECT 
            II.InvoiceId,
            I.CustomerId,
            II.TrackId,
            T.AlbumId
        FROM 
            invoice_items AS II
        JOIN 
            invoices AS I ON II.InvoiceId = I.InvoiceId
        JOIN 
            tracks AS T ON II.TrackId = T.TrackId;
        """
        sales_df = pd.read_sql_query(track_data_query, conn)
        
        # 2. Har bir albomdagi treklar sonini hisoblash
        album_track_counts = sales_df.groupby('AlbumId')['TrackId'].nunique().rename('AlbumTrackCount')
        
        # 3. Har bir buyurtmada sotib olingan treklar sonini hisoblash
        invoice_item_counts = sales_df.groupby(['InvoiceId', 'AlbumId'])['TrackId'].count().rename('PurchasedTrackCount')
        
        # 4. Albom treklar sonini sotib olingan treklar soni bilan solishtirish
        comparison_df = pd.merge(invoice_item_counts.reset_index(), 
                                 album_track_counts.reset_index(), 
                                 on='AlbumId', 
                                 how='left')
        
        # 5. Har bir InvoiceId uchun xarid turi tasnifi
        
        # Agar bitta tranzaksiyada butun albomdagi treklar sotib olingan bo'lsa
        comparison_df['IsAlbum'] = (comparison_df['PurchasedTrackCount'] == comparison_df['AlbumTrackCount'])

        # 6. Mijozlar darajasida tahlil qilish
        
        # Har bir xaridning CustomerId sini qo'shish
        invoice_to_customer = sales_df[['InvoiceId', 'CustomerId']].drop_duplicates()
        final_sales_analysis = pd.merge(comparison_df, invoice_to_customer, on='InvoiceId', how='left')

        
        # Mijozning xarid turini aniqlash:
        # Har bir mijoz uchun IsAlbum ustunidagi barcha TRUE/FALSE qiymatlarini ko'rish
        customer_purchase_types = final_sales_analysis.groupby('CustomerId')['IsAlbum'].agg(['all', 'any'])
        
        # 7. Mijozlarni toifalash
        
        # Faqat Albom sotib olganlar: (all=True) va (any=True)
        album_only_customers = customer_purchase_types[(customer_purchase_types['all'] == True)].shape[0]

        # Faqat Trek sotib olganlar: (all=False) va (any=False)
        track_only_customers = customer_purchase_types[(customer_purchase_types['all'] == False) & (customer_purchase_types['any'] == False)].shape[0]

        # Aralash sotib olganlar (qolgan barchasi)
        mixed_customers = customer_purchase_types.shape[0] - album_only_customers - track_only_customers
        
        # 8. Yakuniy Foizni Hisoblash
        total_customers = sales_df['CustomerId'].nunique()
        
        percentage_track_only = (track_only_customers / total_customers) * 100

        
        print("\n*** Natija ***")
        print(f"Barcha unikal mijozlar soni: {total_customers}")
        print(f"1. Faqat Albom sotib olgan mijozlar: {album_only_customers}")
        print(f"2. Faqat Individual Trek sotib olgan mijozlar: {track_only_customers}")
        print(f"3. Aralash sotib olgan mijozlar: {mixed_customers}")
        print("---------------------------------------------------------")
        print(f" Faqat individual treklarni afzal ko'rgan mijozlar foizi: {percentage_track_only:.2f}%")
        
except Exception as e:
    
    print(f"Muammo chiqdi: {e}")
```
