2. 1.Calculate the total sales for each product.

```python
df2["Totally"] = df2[["Product_A", "Product_B", "Product_C"]].sum(axis = 1)
df2
```

2.2 Find the date with the highest total sales.

```python
df2_max_date = df2[df2["Totally"] == df2["Totally"].max()]["Date"]
print(30*("*"))
print("Max totaly product in df2")
print(df2_max_date)
```

2.3  Calculate the percentage change in sales for each product from the previous day.

```python
df2
```

```python
df2["Product_A_change"] = df2["Product_A"].pct_change()*100 # find change part in percentage and equalled new column
df2["Product_B_change"] = df2["Product_B"].pct_change()*100 # find change part in percentage and equalled new column
df2["Product_C_change"] = df2["Product_C"].pct_change()*100 # find change part in percentage and equalled new column
df2 = df2.round(2) # in df2 float numbers rounded two 
df2
```

2.4 Plot a line chart to visualize the sales trends for each product over time.

```python
df2 = df2.set_index("Date")
```

```python
df2[["Product_A", "Product_B", "Product_C"]].plot()
plt.legend(df2[["Product_A", "Product_B", "Product_C"]])
plt.show()
```

3. Exersice

```python
data3 = {
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Hank', 'Ivy', 'Jack'],
    'Department': ['HR', 'IT', 'Marketing', 'IT', 'Finance', 'HR', 'Marketing', 'IT', 'Finance', 'Marketing'],
    'Salary': [60000, 75000, 65000, 80000, 70000, 72000, 68000, 78000, 69000, 76000],
    'Experience (Years)': [3, 5, 2, 8, 4, 6, 3, 7, 2, 5]
}

df3 = pd.DataFrame(data3)
```

3.1 Calculate the average salary for each department.

```python
df3
```

```python
avg_salary_by_department =  df3.groupby(["Employee_ID", "Department"])["Salary"].mean()
print(f" Avarage Salary each Department: {avg_salary_by_department}")
```

3.2 Find the employee with the most experience.

```python
most_experienced = df3[df3["Experience (Years)"] == df3["Experience (Years)"].max()]
print(90*("*"))
print(f"   Most Experienced Employee is: \n{most_experienced}")
```

3.3 Create a new column 'Salary Increase' representing the percentage increase in salary from the minimum salary in the dataframe.

```python
min_salary = df3["Salary"].min()
min_salary
df3["Salary_increase"] = ((df3["Salary"]/min_salary)*100).round(2)
df3
```

3.4  Plot a bar chart to visualize the distribution of employees across different departments.

```python
employees_by_department = df3.groupby("Department")["Employee_ID"].count()
ax = employees_by_department.plot(kind="bar", figsize = (8,6), title = "Employees by Deprtment")
for i, v in enumerate(employees_by_department):
     ax.text(i, v + 0.2, str(v))

plt.ylabel("Number of Employees")
plt.xlabel("Department")
plt.show()
```
