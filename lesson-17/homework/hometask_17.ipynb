```python
data = {'First Name': ['Alice', 'Bob', 'Charlie', 'David'], 'Age': [25, 30, 35, 40], 'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']} 
df = pd.DataFrame(data)
print(df)
```

```python
def alter_name(name):
    return name.strip().lower().replace(" ","_")

df = df.rename(columns = lambda x: alter_name(x))
print(df)
```

1. Create a DataFrame named sales_and_expenses with columns 'Month', 'Sales', and 'Expenses', representing monthly sales and expenses data. Use below table.

```python
Month = ['Jan', 'Feb','March','April'] # avval ustunlarni list shakliga keltirdim
Sales = [5000, 6000, 7500, 8000] # Sales ni list shakliga keltirdim
Expenses = [3000, 3500, 4000, 4500] # Expenses ni list shakliga keltirdim
my_dict = {"Month": Month, "Sales": Sales, "Expenses": Expenses} # my_dictni yasasim. keyga ustun nomlarini berdim. Valuesga listlarni berdim
df = pd.DataFrame(my_dict) # pd librarydan Dataframe konstruktoridan my_dict dan Dataframe yasab uni df ga tengab oldim
print(df) # result ni print qildim 
```

2.Calculate and display the maximum sales and expenses.

```python
print(df['Sales'].max()) # Sales ustunini Series qilib chqairib max() methodini qolladim 
print(df[['Sales']].max()) # Sales ustunini DF qilib chqairib max() methodini qo'lladim
```

3. Calculate and display the minimum sales and expenses.

```python
print(f"This is Series Object: {df['Sales'].min()}")
print(f"This is DF object: {df[['Sales']].min()}")
```

4. Calculate and display the average sales and expenses.

```python
print(f"This is Avarage Sales {df["Sales"].mean()}\n"
      f"This in avarage of Expenses: {df["Expenses"].mean()}")
```

1. Create a DataFrame named expenses with columns 'Category', 'January', 'February', 'March', and 'April', representing monthly expenses for different categories. Use below table.

```python
Category = ['Rent','Utilities','Groceries','Entertaiment']
January = [1200, 200, 300, 150]
February = [1300, 220, 320, 160]
March = [1400, 240, 330, 170]
April = [1500,250,350, 180]
my_dictik = {'Category': Category, 'January':January, 'February':February, 'March':March, 'April':April}
data = pd.DataFrame(my_dictik)
print(data)
```

2. Calculate and display the maximum expense for each category.

```python
print(50*('*'))
print(f"This is max values grouped by Category\n {data.groupby('Category').max()}") # category buyicha gyuruhlaydi va max qiymatlar return qiladi4
print(50*('*'))
print(f"This is only April max values\n {data.groupby('Category')['April'].max()}")
```

```python
print(50*('*'))
print(f"This is min values grouped by Category\n {data.groupby('Category').min()}") # category buyicha gyuruhlaydi va max qiymatlar return qiladi4
print(50*('*'))
 
```

```python
print(50*('*'))
print(f"This is avarage values grouped by Category\n {data.groupby('Category').mean()}") # category buyicha gyuruhlaydi va max qiymatlar return qiladi4
print(50*('*'))
```
