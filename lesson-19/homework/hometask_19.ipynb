1.2 Sotilgan umumiy miqdorga qarab har bir toifadagi eng ko'p sotiladigan mahsulotni aniqlang.

```python
category  = sales_data.groupby(["Category", "Product"])["Quantity"].sum().reset_index()
# sales_data ni Category va Product boyicha guruhlab olib, shu guruhlar uchun Quantity jamisini hisobladim, indexlarni reset qildim
idx = category.groupby("Category")["Quantity"].idxmax()
#chiqqan natijani yana category boyicha guruhlab, Quantity si boyicha idamax() larini topib oldim va uni idx ga tengladim 
idx
most_sold_products = category.loc[idx]
#Endi dastlabki jadvaldan shu idx qiymarlarni chaqirib oldim 
print(50*('*'))
print("Eng ko'p sotilgan mahsulotlar:")
print(most_sold_products)
print(50*('*'))
# sales_data.groupby("Category").max("Price") 
```

```python
dates = sales_data.groupby(["Category", "Product"])["Date"]
sales_data.head(5)
sales_data["total_sale"] = sales_data["Quantity"] * sales_data["Price"] # umumiy stuvni aniqlab oldim 
sales_data.head(5)
daily_sales = sales_data.groupby("Date")["total_sale"].sum()
daily_sales_index = daily_sales.idxmax()
daily_sales_index 
```

2.2 Bir birlik uchun o'rtacha narxi 120 dollardan yuqori bo'lgan mahsulotlarga buyurtma bergan mijozlarni aniqlang.

```python
product_mean_price = customers.groupby("Product")["Price"].mean()
expensive_products = product_mean_price[product_mean_price> 120].index.tolist()
print(50* ('*'))
print("O'rtacha narxi 120 dan yuqori bolgan mahsulotlar")
print(expensive_products)
```

```python
filtered_product = customers[customers["Product"].isin(expensive_products)]
# customer jadvalini filterlab oldik 
# endi shu mahsulotlarni kim sotib olganani aniqlashimiz kerak
filtered_unique_customer = filtered_product["CustomerID"].unique() # shu mahsulotlarni sotib olgan unikal CustomerID larni topdim 
print("Bir birlik uchun o'rtacha narxi 120 dollardan yuqori bo'lgan mahsulotlarga buyurtma bergan mijozlar")
print(filtered_unique_customer)
```
